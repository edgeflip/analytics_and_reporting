```python
from generate_report import new_hour_queryfrom generate_report import get_campaign_stuff_for_clientfrom time import strftime
```

```python
def generate_hourly_object(client_id):    current_hour = int(strftime('%H'))    obj = {}    mysql_stuff = new_hour_query()    campaign_stuff = get_campaign_stuff_for_client(client_id)    for campaign in campaign_stuff:        obj[campaign[1]] = [[e[1:] for e in mysql_stuff if e[0] == campaign[0]]        for hour in range(0, current_hour):            if hour not in [w[0] for w in obj[campaign[1]]]:                new_list = [hour] + [0 for i in range(9)]                obj[campaign[1]].append(new_list)    return obj
```

```python
obj = generate_hourly_object(2)
```
